# Lectura de datos: PracticasEolica.csv

Este notebook carga y realiza un análisis exploratorio básico del archivo de datos eólicos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Configuración para que los gráficos se vean bien
%matplotlib inline
plt.style.use('ggplot')
```

## Carga de datos

Utilizamos `pd.read_csv` con el separador `;` detectado en la inspección visual del archivo.

```python
file_path = 'PracticasEolica.csv'

try:
    # Intentamos leer con utf-8
    df = pd.read_csv(file_path, sep=';')
except UnicodeDecodeError:
    # Si falla, intentamos con latin-1
    df = pd.read_csv(file_path, sep=';', encoding='latin-1')

# Convertir la columna de Fecha-hora a formato datetime
df['Fecha-hora'] = pd.to_datetime(df['Fecha-hora'], dayfirst=True)

df.head()
```

## Información básica del dataset

```python
df.info()
```

```python
df.describe()
```

## Visualización rápida

Graficamos la velocidad del viento a 10m a lo largo del tiempo.

```python
plt.figure(figsize=(15, 6))
plt.plot(df['Fecha-hora'], df['Velocidad viento 10 m m/s'], color='blue', alpha=0.6)
plt.title('Velocidad del viento a 10 m')
plt.xlabel('Fecha')
plt.ylabel('Velocidad (m/s)')
plt.show()
```
